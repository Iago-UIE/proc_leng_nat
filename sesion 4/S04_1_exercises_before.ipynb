# Tokenization

Split a text into words.

```python
# 1 Exercise: Split the file cantares_gallegos.txt into words. 
with open("cantares_gallegos.txt", "r", encoding="utf-8") as archivo:
    contenido = archivo.read()
print(contenido)


contenido_separado = contenido.split()
# 2 Exercise: Count the number of ocurrences of each word in the file cantares_gallegos.txt.

print(len(contenido_separado))



from collections import Counter
import re

# Normalizar el texto: eliminar puntuación y convertir a minúsculas
palabras = re.findall(r'\b\w+\b', contenido.lower())

# Contar palabras
conteo = Counter(palabras)

# Mostrar resultado
print(conteo)


# 3 Exercise: Order the words by number of ocurrences in the file cantares_gallegos.txt and print the 20 most common words.


# Ordenar palabras por frecuencia
conteo_ordenado = conteo.most_common()

# Mostrar las 20 palabras más comunes
print(conteo_ordenado[:20])

```
